Lecture du fichier `nuage_de_points.txt` en utilisant `readtable` en considérant la première ligne comme noms de colonnes.

```python
data = readtable('nuage_de_points.txt', 'Delimiter', '\t', 'ReadVariableNames', true, 'TextType','string');
```

Conversion des colonnes de `data` en nombres, en remplaçant les virgules par des points.


```python

for col = 1:width(data)
    strCol = string(data{:,col});
    strCol = strrep(strCol, ',', '.');
    data{:,col} = str2double(strCol);
end
```

Extraction des données

```python

x = double(data.X);
y = double(data.Y);
z = double(data.Z);
```

Calcul du plan de moindres carrés et affichage de son equation

```python


mat = [ x y ones(size(x))];
sol = mat\z;
a=sol(1);
b=sol(2);
c=sol(3);
fprintf('Le plan des moindres carrés est : z = %.2f*x + %.2f*y + %.2f\n', a, b, c)
```


Affichage du nuage de points 3D et du plan de moindres carrés.



```python

figure;
scatter3(x, y, z, 25, 'filled');
hold on;
[xg,yg] = meshgrid(linspace(min(x), max(x), 10), linspace(min(y), max(y), 10)); %grid pour avoir un surface smooth
zg = a*xg + b*yg + c;
surf(xg, yg, zg, 'EdgeColor', 'none');
```
